```python
#패키지 임포트
import numpy as np
import pymannkendall as mk # mk분석 패키지
import pandas as pd
import matplotlib.pyplot as plt #시각화 패키지
%matplotlib inline
import matplotlib.font_manager as fm #한글폰트 패키지
# font settong
plt.rcParams['font.family'] = 'HCR Batang'#함초롱바탕
import os #경로 설정시 필요한 패키지
# plt box setting
from matplotlib.offsetbox import AnchoredText 
import scipy.stats as stats
from sklearn.metrics import r2_score
import matplotlib.pylab as pylab
## encoding으로 오류뜰시 euc-kr로 변경해보고 다시 시도
df = pd.read_csv('%s/CSV)Trend + 밭.csv'%(os.getcwd()), encoding='utf-8')
```

```python
df = df.set_index('소규모수도시설명')
df = df.sort_values(by='비율')
```

```python
df
```

```python
rate_data = df['비율']
slope_data = df['Slope']

fig, ax = plt.subplots(figsize = (16,10))
x = rate_data*100
y = slope_data
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
r2 = r2_score(y, p(x))

if (df['분류 기호'].iloc[0] == 'F') == True :
    area_name = '밭 지역'
elif (df['분류 기호'].iloc[0] == 'L') == True :
    area_name = '축산 시설 지역'
elif (df['분류 기호'].iloc[0] == 'N') == True :
    area_name = '자연지 지역'
elif (df['분류 기호'].iloc[0] == 'P') == True :
    area_name = '논 지역'
elif (df['분류 기호'].iloc[0] == 'R') == True :
    area_name = '주거지 지역'
elif (df['분류 기호'].iloc[0] == 'I') == True :
    area_name = '공업지 지역'
else :
    print("알수없는 지역입니다.")

pylab.plot(x,y, 'o', color = 'black')
pylab.plot(x,p(x), "r--")
ax.axhline(y=0, color = 'gray')
ax.axvline(x=0, color = 'gray')
equation ="y=%.6fx+(%.6f)"%(z[0],z[1])

Main_color = 'Black' 
ax.set_title('%s 비율별 그래프'%(area_name), fontsize = 30 ,pad = 20, color = Main_color)
ax.set_xlabel('관정별 250m 버퍼내의 %s 비율(%s)'%(area_name, '%'), fontsize=15, color = Main_color)
ax.set_ylabel('Sens Slope Value', fontsize = 15, color = Main_color)
ax.tick_params(axis = 'x', direction='in', length = 10, labelcolor = Main_color, color = Main_color)
ax.tick_params(axis = 'y', direction='in', labelcolor = Main_color, color = Main_color)

text = "y = %.6fx + (%.6f)\nR^2 = %.6f" % (z[0], z[1], r2)
ax.text(x.max()*0.7, z[0]*x.max()+z[1], text, fontsize=14, verticalalignment='top')
fig.savefig('%s.png'%(area_name), dpi=150, facecolor='#eeeeee', bbox_inches='tight')

plt.show()
```
